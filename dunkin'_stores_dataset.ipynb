# **Cleaning:**

```python
df_new = df[['row_id','city', 'state','address_line_1', 'address_line_2',
       'phone', 'mon_hrs', 'tue_hrs', 'wed_hrs', 'thu_hrs', 'fri_hrs',
       'sat_hrs', 'sun_hrs', 'drive-thru', 'mobile-order', 'dunkin-card',
       'kcup', 'curbside-pickup', 'has-baskin-robbins']]
df_new.head(3)
```

```python
df_new.isna().sum()
```

```python
df_new[df_new['phone'].isna()]
```

```python
df_new.dtypes
```

```python
df_new[df_new['address_line_2'].isna()]
```

```python
df_new.drop('address_line_2', inplace=True, axis=1)
df_new.head(2)
```

```python
df_new["phone"].fillna('0123456789', inplace = True)
```

```python
df_new.isna().sum()
```

```python
df_new.drop('phone', inplace=True, axis='columns')
df_new.head(2)
```

```python
df_new=df_new.rename(columns={'address_line_1':'address'})
df_new.head()
```

```python
print(df_new.duplicated())
```

```python
df_new.replace({False:0,True:1},inplace=True)

df_new.head()
```
