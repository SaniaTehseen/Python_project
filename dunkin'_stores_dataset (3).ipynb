# **Cleaning:**

```python
df_new = df[['row_id','city', 'state','address_line_1', 'address_line_2',
       'phone', 'mon_hrs', 'tue_hrs', 'wed_hrs', 'thu_hrs', 'fri_hrs',
       'sat_hrs', 'sun_hrs', 'drive-thru', 'mobile-order', 'dunkin-card',
       'kcup', 'curbside-pickup', 'has-baskin-robbins']]
df_new.head(3)
```

```python
df_new.isna().sum()
```

```python
df_new[df_new['phone'].isna()]
```

```python
df_new.dtypes
```

```python
df_new[df_new['address_line_2'].isna()]
```

```python
df_new.drop('address_line_2', inplace=True, axis=1)
df_new.head(2)
```

```python
df_new["phone"].fillna('0123456789', inplace = True)
```

```python
df_new.isna().sum()
```

```python
df_new.drop('phone', inplace=True, axis='columns')
df_new.head(2)
```

```python
df_new=df_new.rename(columns={'address_line_1':'address'})
df_new.head()
```

```python
print(df_new.duplicated())
```

```python
df_new.replace({False:0,True:1},inplace=True)

df_new.head()
```

```python
# split column and add new columns to df_new
df_new[['mon_opening_hrs', 'mon_closing_hrs']] = df_new['mon_hrs'].str.split('-', expand=True)
df_new[['tue_opening_hrs', 'tue_closing_hrs']] = df_new['tue_hrs'].str.split('-', expand=True)
df_new[['wed_opening_hrs', 'wed_closing_hrs']] = df_new['wed_hrs'].str.split('-', expand=True)
df_new[['thu_opening_hrs', 'thu_closing_hrs']] = df_new['thu_hrs'].str.split('-', expand=True)
df_new[['fri_opening_hrs', 'fri_closing_hrs']] = df_new['fri_hrs'].str.split('-', expand=True)
df_new[['sat_opening_hrs', 'sat_closing_hrs']] = df_new['sat_hrs'].str.split('-', expand=True)
df_new[['sun_opening_hrs', 'sun_closing_hrs']] = df_new['sun_hrs'].str.split('-', expand=True)

# display the dataframe
df_new
```

```python
df_new.columns
```

```python
dunkin_dataset=df_new[['row_id','city', 'state','address',
        'mon_opening_hrs', 'mon_closing_hrs', 'tue_opening_hrs',
       'tue_closing_hrs', 'wed_opening_hrs', 'wed_closing_hrs',
       'thu_opening_hrs', 'thu_closing_hrs', 'fri_opening_hrs',
       'fri_closing_hrs', 'sat_opening_hrs', 'sat_closing_hrs',
       'sun_opening_hrs', 'sun_closing_hrs', 'drive-thru', 'mobile-order', 'dunkin-card',
       'kcup', 'curbside-pickup', 'has-baskin-robbins']]
pd.set_option('display.max_columns', None)
dunkin_dataset.head()       
```

```python
final_data_set = dunkin_dataset[['row_id','city', 'state','address',
        'drive-thru', 'mobile-order', 'dunkin-card',
       'kcup', 'curbside-pickup', 'has-baskin-robbins']]

final_data_set.head()       
```

```python
item_counts=final_data_set['state'].value_counts()
print(item_counts)
```

```python
clean_states={'state':{'NY':1,"MA":2,'FL':3,"NJ":4,'IL':5,'PA':6,'CT':7,'MD':8,'GA':9,'OH':9,'VA':10,'NH':11 ,'NC':12,'TX':13,'RI':14,'ME':15,'CA':16,'TN':17,'AZ':18,'SC':19,
                        'IN':20,"MI":21,'WI':22,'DE':23,'AL':24,'MO':25,'VT':26,'CO':27,'KY':28,'NV':29,'IA':30,'MN':31,'KS':32,"WV":33,'NE':34,'DC':35,'NM':36,'OK':37,'LA':38,
                        'MS':41,"AR":42,"HI":43,'AK':44,'UT':45,'WY':46}}
```

```python
final_data_set=final_data_set.replace(clean_states)
final_data_set.head()
```

```python
item_counts=final_data_set['address'].value_counts()
print(item_counts)
```
